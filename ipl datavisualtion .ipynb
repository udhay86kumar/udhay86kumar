```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

```

```python
data = pd.read_csv("C:/Users/udhaya/Desktop/IPL Matches 2008-2020.csv") 
```

```python
data.head()
```

```python
data.info()
```

```python
temp= pd.DataFrame({"Winner": data['winner']})
count_wins = temp.value_counts()
#print(count_wins)
labels=[X[0] for X in count_wins.keys()]
bar , ax = plt.subplots(figsize=(20,12))

ax = plt.pie(x = count_wins, labels=labels)
plt.title("Most wins in IPL",fontsize =25)
```

```python
sns.countplot(data['winner'][data['eliminator']=='Y'],data=data)
plt.title("Most win in Eliminator ", fontsize = 17)

plt.xticks(rotation=90)
plt.show()
```
