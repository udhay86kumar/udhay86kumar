```python
from sklearn import datasets
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis 
```

```python
iris = datasets.load_iris()
x=iris.data
y=iris.target
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)
```

```python
x_lda=lda.fit(x,y).transform(x)
```
